# 1я часть RE

```python
import urllib.request

url = "https://en.wikipedia.org/wiki/Category:American_nuclear_physicists"
headers ={ "User-Agent" : "Mozilla/5.0" }
req = urllib.request.Request(url, None, headers)
fp = urllib.request.urlopen(req)

text = fp.read()
text = text.decode("utf8")

fp.close()
```

```python
htmlFile = open("HTML.html", "w")

htmlFile.write(text)


htmlFile.close()
```

```python
import re
```

```python
scientists_groups = re.findall(r"(<h3>[A-Z]{1}</h3>)((?:(?!</div>).|\n)*)</div>", text)
```

```python
def ExtractScientists(strings):
    scientists = []
    scientists_links = []
    for string in strings:
        scientists.append("https://en.wikipedia.org/" + re.findall(r"href=\"((?:(?!\").)*)\"", string)[0])
        scientists_links.append(re.findall(r"title=\"((?:(?!\").)*)\"", string)[0])
        
    return scientists_links, scientists
```

```python
scientists = []
scientists_links = []
for group_ in scientists_groups:
    group = group_[1]
    
#     print('A')
#     print(re.findall(r"<li>((?:(?!</li>).|\n)*)</li>", group))
#     print('B')
    
    scientists_group, scientists_group_link = ExtractScientists(re.findall(r"<li>((?:(?!</li>).|\n)*)</li>", group))
    
    scientists += scientists_group
    scientists_links += scientists_group_link
    
    
#     print('A')
#     print(group)
#     print('B')
#     for string in group:
#         print('A')
#         print(string)
#         print('B')
```

```python
n = len(scientists_links)
for i in range(n):
    scientists_links[i] += "\n"
    scientists[i] += "\n" 
fp_names = open("ScientistsLinks.txt", 'w')
fp_names.writelines(scientists_links)
fp_names.close()

fp_names = open("Scientists.txt", 'w')
fp_names.writelines(scientists)
fp_names.close()
```

# 2я часть RE

```python
def GetInstitutions(url):
    institutions = []
    headers ={ "User-Agent" : "Mozilla/5.0" }
    req = urllib.request.Request(url, None, headers)
    fp = urllib.request.urlopen(req)
    text = fp.read()
    text = text.decode("utf8")
    fp.close()
    
    res = re.findall(r"infobox biography vcard((?:(?!</table>).|\n)*)</table>", text)
    if len(res):
        res = re.findall(r"Institutions((?:(?!</td>).|\n)*)</td>", res[0])
        if len(res):
            return re.findall(r"title=\"((?:(?!\").)*)\"", res[0])
    
#     res = re.findall(r"title=\"((?:(?!\").)*)\"", res)
    return res
```

```python
institutions = {}
qua_of_inst = 0
related_inst = []

for name, link in zip(scientists, scientists_links):
    print(name, link)
    tmp_institutions = GetInstitutions(link)
    for curr_inst in tmp_institutions:
#       add institutions
        if institutions.get(curr_inst) is None:
            qua_of_inst += 1
            institutions[curr_inst] = qua_of_inst
        
#   add relations    
    curr_n = len(tmp_institutions)
    for i in range(curr_n):
        for j in range(curr_n):
            if i != j:
                related_inst.append((tmp_institutions[i], tmp_institutions[j], name))
        
    
```

```python
with open("InstituteRelations.tgf", "w") as fp:
    for institute in institutions:
        fp.write(str(institutions[institute]) + " " + str(institute) + "\n")
    
    fp.write("#\n")
    
    for relation in related_inst:
        fp.write(str(institutions[relation[0]]) + " " + str(institutions[relation[1]]) + " " + str(relation[2]) + "\n")
          
    
```
